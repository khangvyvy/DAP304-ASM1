```python
import re
import statistics

# Tạo hàm tìm câu hỏi được bỏ qua nhiều nhất
def find_most_skipped(skipped_questions,line_count):
    max_skipped = max(skipped_questions)
    most_skipped = []
    for i, skipped in enumerate(skipped_questions):
        if skipped == max_skipped:
            most_skipped.append((i+1, skipped_questions[i], round(skipped_questions[i] / line_count, 2)))
    return most_skipped

# Hàm tìm câu hỏi có số lượng câu trả lời sai nhiều nhất
def find_most_wrong(wrong_answers,line_count):
    max_wrong = max(wrong_answers)
    most_wrong = []
    for i, wrong in enumerate(wrong_answers):
        if wrong == max_wrong:
            most_wrong.append((i+1, wrong_answers[i], round(wrong_answers[i] / line_count, 2)))
    return most_wrong
    
# Hàm phân tích bài thi
def analyze_file(file_name):
    try: 
        with open(file_name,"r") as file:
            print("Đã mở được file {0}".format(file_name))
            # Gán biến mô hình hợp lệ, các biến đếm, biến câu trả lời
            pattern1= r'^N[0-9]{8}'
            valid=0
            invalid=0
            line_count = 0
            score=0
            score_list=list()
            
            answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
            answer_list=answer_key.strip().split(',')
            skipped_questions = [0] * len(answer_key)
            wrong_answers = [0] * len(answer_key)
            
            # Lặp qua từng dòng trong file
            print("\nPHÂN TÍCH")
            for line in file:
                line_count = line_count+1
                student_answer=line.strip().split(',')
                
                # Kiểm tra mã học sinh và số lượng câu trả lời
                if re.match(pattern1, student_answer[0]) and len(student_answer)==26:
                    valid=valid+1
                    # Bỏ mã học sinh và xử lý câu trả lời        
                    student_answer.pop(0)
                    for i, (correct_ans, student_ans) in enumerate(zip(answer_list, student_answer)):
                        if student_ans == '':
                            skipped_questions[i] += 1
                        elif student_ans == correct_ans:
                            score += 4
                        else:
                            score -= 1
                            wrong_answers[i] += 1
                    score_list.append(score)
                    score=0
                else:
                    invalid=invalid+1
                # In ra nếu dòng không hợp lệ lý do là mã học sinh không đúng quy cách
                    if not(re.match(pattern1, student_answer[0])):
                        print("Không hợp lệ: Mã học sinh không đúng\n{0}".format(line))
                    # In ra nếu dòng không hợp lệ lý do là không đủ 26 giá trị trong câu trả lơi
                    else:
                        print("Không hợp lệ: Không đủ câu trả lời\n{0}".format(line)) 

            high_score= sum(score > 80 for score in score_list)
            mean=round(statistics.mean(score_list), 2)
            highest=max(score_list)
            lowest=min(score_list)
            range_score=highest-lowest
            median=round(statistics.median(score_list), 3)
            
             # In ra kết quả phân tích
            print("\nKẾT QUẢ PHÂN TÍCH")
            print("Tổng số học sinh: ",line_count)
            print("Số bài hợp lệ: ",valid)
            print("Số bài không hợp lệ: ",invalid)
            print("Số học sinh có điểm cao trên 80: ",high_score)
            print("Điểm trung bình của các học sinh: ", mean)
            print("Điểm cao nhất: ",highest)
            print("Điểm thấp nhất: ",lowest)
            print("Khoảng cách giữa điểm cao nhất và thấp nhất: ",range_score)
            print("Điểm trung vị: ",median)
            print("Các câu hỏi học sinh bỏ qua nhiều nhất:", find_most_skipped(skipped_questions,(line_count-invalid))) 
            print("Các câu hỏi học sinh làm sai nhiều nhất:",find_most_wrong(wrong_answers,(line_count-invalid)))
            
    # Thông báo nếu tên file không tìm thấy
    except FileNotFoundError:
        print("Không tìm thấy file {0}, vui lòng thử lại.".format(file_name))
        find_file()
            
# Hàm để người dùng input và đọc hàm read_file
def find_file():
    file_name=input("Nhập tên file cần đọc:")
    analyze_file(file_name)
    
# Để chạy trực tiếp hàm find_file mà không cần gọi
if __name__ == "__main__":
    find_file()
```

```python
import pandas as pd
import re
import statistics

def find_most_skipped(skipped_questions, line_count):
    max_skipped = max(skipped_questions)
    most_skipped = []
    for i, skipped in enumerate(skipped_questions):
        if skipped == max_skipped:
            most_skipped.append((i+1, skipped_questions[i], round(skipped_questions[i] / line_count, 2)))
    return most_skipped

def find_most_wrong(wrong_answers, line_count):
    max_wrong = max(wrong_answers)
    most_wrong = []
    for i, wrong in enumerate(wrong_answers):
        if wrong == max_wrong:
            most_wrong.append((i+1, wrong_answers[i], round(wrong_answers[i] / line_count, 2)))
    return most_wrong
    
def analyze_file(file_name):
    try:
        df = pd.read_csv(file_name, header=None)
        print("Đã mở file thành công:", file_name)

        pattern1 = r'^N[0-9]{8}'
        valid = 0
        invalid = 0
        score_list = []
        skipped_questions = [0] * 26
        wrong_answers = [0] * 26
        student_ids = []

        answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
        answer_list = answer_key.strip().split(',')

        for idx, row in df.iterrows():
            student_id = row[0]
            student_answer = row[1:]

            if re.match(pattern1, student_id) and len(student_answer) == 26:
                valid += 1
                student_ids.append(student_id)
                score = 0
                for i, (correct_ans, ans) in enumerate(zip(answer_list, student_answer)):
                    if pd.isnull(ans):
                        skipped_questions[i] += 1
                    elif ans == correct_ans:
                        score += 4
                    else:
                        score -= 1
                        wrong_answers[i] += 1
                score_list.append(score)
            else:
                invalid += 1
                if not re.match(pattern1, student_id):
                    print("Dòng dữ liệu không hợp lệ: N# không hợp lệ\n{0}".format(student_id))
                else:
                    print("Dòng dữ liệu không hợp lệ: không chứa đúng 26 giá trị:\n{0}".format(student_id))

        line_count = len(df)
        high_score = sum(score > 80 for score in score_list)
        mean = round(statistics.mean(score_list), 2)
        highest = max(score_list)
        lowest = min(score_list)
        range_score = highest - lowest
        median = round(statistics.median(score_list), 3)

        print("\nKẾT QUẢ PHÂN TÍCH")
        print("Tổng số học sinh:", line_count)
        print("Số hồ sơ hợp lệ:", valid)
        print("Số hồ sơ không hợp lệ:", invalid)
        print("Học sinh có điểm cao hơn 80:", high_score)
        print("Điểm trung bình của học sinh:", mean)
        print("Điểm cao nhất:", highest)
        print("Điểm thấp nhất:", lowest)
        print("Khoảng điểm:", range_score)
        print("Điểm trung vị:", median)
        print("Câu hỏi bị bỏ qua nhiều nhất:", find_most_skipped(skipped_questions, line_count - invalid))
        print("Câu hỏi bị trả lời sai nhiều nhất:", find_most_wrong(wrong_answers, line_count - invalid))

        result_df = pd.DataFrame({'Mã học sinh': student_ids, 'Điểm': score_list})
        result_file_name = file_name.split('.')[0] + '_grades.txt'
        result_df.to_csv(result_file_name, index=False)
        print("Kết quả chi tiết đã được lưu trong:", result_file_name)

    except FileNotFoundError:
        print("Không tìm thấy file:", file_name)

def find_file():
    file_name = input("Nhập tên file cần phân tích: ")
    analyze_file(file_name)

if __name__ == "__main__":
    find_file()
```
